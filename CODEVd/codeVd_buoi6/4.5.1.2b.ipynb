```python
import pandas as pd
import numpy as np

# 1. Tạo DataFrame có cột toàn NA
df_demo = pd.DataFrame({
    'A': [1, 2, np.nan, 4],
    'B': [np.nan, np.nan, np.nan, np.nan],
    'C': [5.0, np.nan, 8.0, 9.0]
})
print("DataFrame có cột toàn NA:")
print(df_demo)

# 2. Sao chép DataFrame để tránh thay đổi dữ liệu gốc
df_all_na_col = df_demo.copy()

# 3. Xóa các cột (axis=1) mà TOÀN BỘ giá trị là NaN (how='all')
df_xoa_col = df_all_na_col.dropna(axis='columns', how='all')
print("\nKết quả dropna(axis='columns', how='all'):")
print(df_xoa_col)

# 4. So sánh lại DataFrame sau khi xóa
print("\nCác cột bị xóa là cột nào có toàn bộ giá trị NaN.")
```
