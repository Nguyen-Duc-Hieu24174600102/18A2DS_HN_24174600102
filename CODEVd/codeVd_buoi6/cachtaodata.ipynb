```python
import pandas as pd
import numpy as np

# ------------------------------
# C1. TẠO DataFrame TỪ MỘT SERIES
# ------------------------------
print("===== C1. Tạo DataFrame từ Series =====")
population = pd.Series(
    [8330830, 1195490, 9411805, 1836270],
    index=['Hà Nội', 'Đà Nẵng', 'TP. Hồ Chí Minh', 'Nam Định']
)

df1 = pd.DataFrame(population, columns=['Dân số'])
print(df1, "\n")

# ------------------------------
# C2. TẠO DataFrame TỪ LIST CỦA DICTIONARY
# ------------------------------
print("===== C2. Tạo DataFrame từ List của Dictionary =====")
data = [{'a': i, 'b': 2 * i} for i in range(3)]
print("Danh sách dict:", data)

df2 = pd.DataFrame(data)
print("\nDataFrame tạo từ List của Dictionary:")
print(df2, "\n")

# Khi các khóa bị thiếu, pandas tự động thêm NaN
df3 = pd.DataFrame([{'a': 1, 'b': 2}, {'b': 3, 'c': 'KHDL'}])
print("DataFrame với khóa bị thiếu:")
print(df3, "\n")

# ------------------------------
# C3. TẠO DataFrame TỪ DICTIONARY CỦA SERIES
# ------------------------------
print("===== C3. Tạo DataFrame từ Dictionary chứa các Series =====")
population = pd.Series(
    [8330830, 1195490, 9411805, 1836270],
    index=['Hà Nội', 'Đà Nẵng', 'TP. Hồ Chí Minh', 'Nam Định']
)
area = pd.Series(
    [3358, 1285, 2061, 1668],
    index=['Hà Nội', 'Đà Nẵng', 'TP. Hồ Chí Minh', 'Nam Định']
)

VietNam = pd.DataFrame({'Dân số': population, 'Diện tích': area})
print(VietNam, "\n")

# ------------------------------
# C4. TẠO DataFrame TỪ MẢNG 2 CHIỀU NUMPY
# ------------------------------
print("===== C4. Tạo DataFrame từ mảng 2 chiều NumPy =====")
arr = np.random.rand(3, 2)
print("Mảng NumPy:\n", arr)

df4 = pd.DataFrame(arr, columns=['foo', 'bar'], index=['a', 'b', 'c'])
print("\nDataFrame tạo từ NumPy array:")
print(df4, "\n")

# ------------------------------
# C5. TẠO DataFrame TỪ STRUCTURED ARRAY CỦA NUMPY
# ------------------------------
print("===== C5. Tạo DataFrame từ Structured Array của NumPy =====")
dtype = [('name', 'U20'), ('age', int), ('height', float)]

data = np.array([
    ('Minh', 22, 165.5),
    ('Lan', 40, 160.0),
    ('Linh', 31, 165.0),
    ('Lương', 18, 166.5),
    ('Thanh', 50, 170.0)
], dtype=dtype)

df5 = pd.DataFrame(data)
print(df5)
print("\nKiểu dữ liệu Structured Array:", data.dtype)
```
