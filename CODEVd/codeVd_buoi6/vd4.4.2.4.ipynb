```python
import pandas as pd

# Khởi tạo một dictionary chứa dữ liệu
data = {
    "Tỉnh/TP": ["Hà Nội", "Đà Nẵng", "TP. Hồ Chí Minh", "Nam Định"],
    "Dân số (người)": [8330830, 1195490, 9411805, 1836270],
    "Diện tích (km2)": [3358, 1285, 2061, 1668]
}

# Tạo đối tượng DataFrame từ dictionary
df = pd.DataFrame(data)

# In DataFrame ra màn hình
print(df)

#Lấy dữ liệu dân số và mật độ các tỉnh/TP có diện tích > 2000km2.   
print(data.loc[data['Diện tích'] > 2000, ['Dân số', 'Mật độ']]) 
print("Diện tích các tỉnh/TP có mật độ < 1000 và dân số > 1 triệu:\n",data.loc[(
 data['Mật độ']<1000) & (data['Dân số'] > 1000000), ['Diện tích', 'Mật độ']
 ])
```
