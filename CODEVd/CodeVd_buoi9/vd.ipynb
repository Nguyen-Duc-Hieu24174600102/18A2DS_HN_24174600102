```python
import threading
import time

sem = threading.Semaphore(3)

def access(id):
    sem.acquire()
    print(f"Thread {id} bắt đầu truy cập")
    time.sleep(1)
    print(f"Thread {id} kết thúc")
    sem.release()

for i in range(6):
    threading.Thread(target=access, args=(i,)).start()
```

```python
import threading
import time

# Tạo một Lock
lock = threading.Lock()
balance = 100

def add_money_safe():
    global balance
    # Yêu cầu khóa. Chỉ 1 luồng được đi qua.
    lock.acquire()
    try:
        tmp = balance            # Vùng mã CS
        time.sleep(0.001)
        balance = tmp + 10
    finally:
        # Nhả khóa, cho phép luồng khác vào
        lock.release()

# Tạo 2 thread, cả 2 đều chạy cùng một hàm add_money_safe()
threads = [threading.Thread(target=add_money_safe) for _ in range(2)]

for t in threads:
    t.start()          # Cả 2 thread bắt đầu gần như cùng lúc

for t in threads:
    t.join()           # Đợi cả 2 thread chạy xong

print("Số dư cuối:", balance)
```
