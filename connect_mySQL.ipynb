```python
import mysql.connector
from mysql.connector import Error

# Cấu hình kết nối
PORT = 3307
USER = "root"
PASSWORD = "uneti"
DATABASE = "data_science_db"
HOST = "127.0.0.1"

try:
    conn = mysql.connector.connect(
        host=HOST,
        user=USER,
        password=PASSWORD,
        database=DATABASE,
        port=PORT
    )

    if conn.is_connected():
        print("Kết nối MySQL thành công!")

        cursor = conn.cursor()
        cursor.execute("SELECT DATABASE();")
        db = cursor.fetchone()
        print(f"Đang sử dụng cơ sở dữ liệu: {db[0]}")

except Error as e:
    print(f"Lỗi khi kết nối MySQL: {e}")

finally:
    if 'conn' in locals() and conn.is_connected():
        cursor.close()
        conn.close()
        print("Đã đóng kết nối MySQL.")
```
