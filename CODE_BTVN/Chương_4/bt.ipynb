```python
# Bài 4.5
ser1 = pd.Series([1, 2, 3, 4, 5])
ser2 = pd.Series([4, 5, 6, 7, 8])

ket_qua = ser1[~ser1.isin(ser2)]
print(ket_qua)
```

```python
# Bài 4.6
ser1 = pd.Series([1, 2, 3, 4, 5])
ser2 = pd.Series([4, 5, 6, 7, 8])

ser_u = pd.Series(np.union1d(ser1, ser2)) # Hợp
ser_i = pd.Series(np.intersect1d(ser1, ser2)) # Giao
ket_qua = ser_u[~ser_u.isin(ser_i)]
print(ket_qua)
```

```python
# Bài 4.7
state = np.random.RandomState(100)
ser = pd.Series(state.normal(10, 5, 25))

# Cách 1: Dùng describe
print(ser.describe())

# Cách 2: Dùng numpy percentile
print(np.percentile(ser, q=[0, 25, 50, 75, 100]))

# Bài 4.8
np.random.seed(10)
ser = pd.Series(np.random.randint(1, 5, [12]))
print("Gốc:", ser.values)

top_2 = ser.value_counts().index[:2]
ser[~ser.isin(top_2)] = 'Other'
print("Kết quả:\n", ser)

# Bài 4.9
ser = pd.Series(np.random.random(20))
ser_deciles = pd.qcut(ser, q=10, labels=False)
print(ser_deciles)

# Bài 4.10
ser = pd.Series(np.random.randint(1, 10, 35))
df = pd.DataFrame(ser.values.reshape(7, 5))
print(df)

# Bài 4.11
ser = pd.Series(np.random.randint(1, 10, 7))
print("Series:", ser.values)

vi_tri = np.argwhere(ser.values % 3 == 0)
print("Vị trí bội số của 3:", vi_tri.flatten())

# Bài 4.12
ser = pd.Series(list('abcdefghijklmnopqrstuvwxyz'))
pos = [0, 4, 8, 14, 20]

ket_qua = ser.take(pos)
print(ket_qua)

# Bài 4.13
ser1 = pd.Series(range(5))
ser2 = pd.Series(list('abcde'))

# Dọc
doc = pd.concat([ser1, ser2], axis=0)
# Ngang
ngang = pd.concat([ser1, ser2], axis=1)

print("Dọc:\n", doc.head())
print("Ngang:\n", ngang.head())

# Bài 4.14
ser1 = pd.Series([10, 9, 6, 5, 3, 1, 12, 8, 13])
ser2 = pd.Series([1, 3, 10, 13])

# Tìm vị trí (index) trong ser1
vi_tri = [pd.Index(ser1).get_loc(i) for i in ser2]
print("Vị trí:", vi_tri)

# Bài 4.15
truth = pd.Series(range(10))
pred = pd.Series(range(10)) + np.random.random(10)

mse = np.mean((truth - pred)**2)
print("MSE =", mse)

# Bài 4.16
ser = pd.Series(['how', 'to', 'kick', 'ass?'])
# Dùng hàm title() hoặc capitalize()
ket_qua = ser.map(lambda x: x.title())
print(ket_qua)

# Bài 4.17 (Đếm ký tự)
ser = pd.Series(['how', 'to', 'kick', 'ass?'])
do_dai = ser.map(len)
print(do_dai)

# Bài 4.18
import re
emails = pd.Series(['buying books at amazon.com', 'rameses@egypt.com', 'matt@t.co', 'narendra@modi.com'])
pattern = r'[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,4}'

valid_emails = emails.str.findall(pattern, flags=re.IGNORECASE)
print([x[0] for x in valid_emails if x])

# Bài 4.19
fruit = pd.Series(np.random.choice(['apple', 'banana', 'carrot'], 10))
weights = pd.Series(np.linspace(1, 10, 10))

mean_weights = weights.groupby(fruit).mean()
print(mean_weights)

# Bài 4.20
# Load dữ liệu
url = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/02_Filtering_%26_Sorting/Euro12/Euro_2012_stats_TEAM.csv"
euro12 = pd.read_csv(url)

# 1. In cột Goals
print("1. Goals:", euro12['Goals'].head().tolist())

# 2. Số đội tham gia
print("2. Số đội:", euro12.shape[0])

# 3. Thông tin cột
print("3. Số cột:", euro12.shape[1])

# 4. Tạo DF discipline
discipline = euro12[['Team', 'Yellow Cards', 'Red Cards']]

# 5. Sắp xếp giảm dần
discipline = discipline.sort_values(['Red Cards', 'Yellow Cards'], ascending=False)
print("5. Top thẻ phạt:\n", discipline.head(3))

# 6. Trung bình thẻ vàng
print("6. TB thẻ vàng:", discipline['Yellow Cards'].mean())

# 7. Đội > 6 bàn thắng
print("7. Đội > 6 bàn:", euro12[euro12.Goals > 6].Team.tolist())

# 8. Đội bắt đầu bằng 'G'
print("8. Đội tên G:", euro12[euro12.Team.str.startswith('G')].Team.tolist())

# 9. 7 cột đầu
print("9. 7 cột đầu:\n", euro12.iloc[:, :7].head(2))

# 10. Trừ 3 cột cuối
print("10. Trừ 3 cột cuối:\n", euro12.iloc[:, :-3].head(2))

# 11. Các cột cụ thể của Eng, Ita, Rus
target = euro12.loc[euro12.Team.isin(['England', 'Italy', 'Russia']), ['Team', 'Shooting Accuracy']]
print("11. Shooting Accuracy:\n", target)

# Bài 4.21
# Load dữ liệu
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv"
drink = pd.read_csv(url)

# 1. Type và Shape
print(f"1. Type: {type(drink)}, Shape: {drink.shape}")

# 2. Bia trung bình mỗi châu lục
print("2. Bia TB:\n", drink.groupby('continent')['beer_servings'].mean())

# 3. Thống kê rượu vang
print("3. Stats Wine:\n", drink.groupby('continent')['wine_servings'].describe())

# 4. Trung bình tất cả các cột
print("4. Mean All:\n", drink.groupby('continent').mean(numeric_only=True))

# 5. Trung vị
print("5. Median:\n", drink.groupby('continent').median(numeric_only=True))

# 6. Rượu mạnh (mean, max, min)
print("6. Spirit Stats:\n", drink.groupby('continent')['spirit_servings'].agg(['mean', 'max', 'min']))

# 7. 5 quốc gia tiêu thụ bia nhiều nhất
print("7. Top 5 Beer:\n", drink.sort_values('beer_servings', ascending=False).head(5)[['country', 'beer_servings']])

# 7. 5 quốc gia tiêu thụ bia ít nhất
print("7. Bottom 5 Beer:\n", drink.sort_values('beer_servings', ascending=True).head(5)[['country', 'beer_servings']])
```
