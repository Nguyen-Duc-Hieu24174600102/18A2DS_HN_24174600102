```python
import numpy as np

# 1. arr từ 0 đến 9
arr = np.arange(10)
print("arr =", arr)
print("dtype =", arr.dtype)
print("shape =", arr.shape)

# 2. arr_odd (lẻ) và arr_even (chẵn)
arr_odd = arr[arr % 2 == 1]
arr_even = arr[arr % 2 == 0]
print("Số lẻ:", arr_odd)
print("Số chẵn:", arr_even)

# 3. thay phần tử chẵn = 100
arr_update = arr.copy()
arr_update[arr_update % 2 == 0] = 100
print("Sau khi thay chẵn = 100:", arr_update)
```
