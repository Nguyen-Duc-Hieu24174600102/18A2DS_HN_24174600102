```python
import numpy as np

# dữ liệu ví dụ
height_in = np.array([72, 70, 68, 75, 78])  # inches
weight_lb = np.array([180, 190, 165, 210, 220])  # pounds

# đổi đơn vị
height_m = height_in * 0.0254
weight_kg = weight_lb * 0.453592

BMI = weight_kg / (height_m ** 2)

print("BMI =", BMI)
print("BMI > 25:", BMI[BMI > 25])
print("Chỉ số lớn nhất:", BMI.max())
print("Vị trí chỉ số lớn nhất:", BMI.argmax())
```
