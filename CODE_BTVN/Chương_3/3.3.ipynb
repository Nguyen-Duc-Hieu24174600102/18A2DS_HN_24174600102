```python
import numpy as np

arr_a = np.array([1,2,2,3,3,4,3,4,5,6])
arr_b = np.array([7,2,10,2,7,4,9,9,8])

# 1. phần tử xuất hiện ở cả 2 mảng
arr_c = np.intersect1d(arr_a, arr_b)
print("Xuất hiện ở cả hai:", arr_c)

# 2. phần tử ở b nhưng không ở a
arr_d = np.setdiff1d(arr_b, arr_a)
print("Có ở b không có ở a:", arr_d)

# 3. chọn phần tử 5–10
err_e = np.array([2,1,9,10,3,27,8,6,25,16])
c = err_e[(err_e >= 5) & (err_e <= 10)]
print("Trong khoảng 5–10:", c)
```
