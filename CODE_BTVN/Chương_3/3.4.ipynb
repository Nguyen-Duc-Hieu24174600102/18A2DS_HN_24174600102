```python
import numpy as np

# 1. mảng zero, thay phần tử thứ 5 = 1
arr_zeros = np.zeros(10)
arr_zeros[4] = 1
print("Câu 1:", arr_zeros)

# 2. mảng từ 10–24, đảo ngược
arr_h = np.arange(10,25)
print("Câu 2 (đảo):", arr_h[::-1])

# 3. lọc phần tử khác 0
arr_k = np.array([1,2,0,0,3,4,0,3,1,0,3,5,0])
arr_l = arr_k[arr_k != 0]
print("Câu 3:", arr_l)

# 4. thay 2 phần tử đầu = 10, 20
arr_l[:2] = [10, 20]
print("Câu 4:", arr_l)

# 5. chèn 100 vào index 5
arr_l = np.insert(arr_l, 5, 100)
print("Câu 5:", arr_l)

# 6. xóa index 0,1,2
arr_l = np.delete(arr_l, [0,1,2])
print("Câu 6:", arr_l)
```
