```python
import numpy as np

baseball = np.array([
    [72, 180],
    [70, 190],
    [68, 165],
    [75, 210],
    [78, 220]
])

# 1. xem kiểu & kích thước
print(baseball.dtype, baseball.shape)

# 2. lấy 50 dòng đầu (ở đây ví dụ)
print(baseball[:50])

# 3. height là cột 0
height = baseball[:,0]

# 4. weight là cột 1
weight = baseball[:,1]

# 5. đổi đơn vị sang mét và kg
height_m = height * 0.0254
weight_kg = weight * 0.453592

# 6. BMI
BMI = weight_kg / height_m**2
print("BMI:", BMI)

# 7. BMI > trung bình
print("BMI lớn hơn trung bình:", BMI[BMI > BMI.mean()])
```
