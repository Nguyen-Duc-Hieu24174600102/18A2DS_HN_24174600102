```python
# code for update operation
import sqlite3

# Kết nối tới cơ sở dữ liệu
conn = sqlite3.connect('mydatabase.db')

# Tạo con trỏ để thực thi câu lệnh SQL
cursor = conn.cursor()

# Cập nhật bản ghi trong bảng users
cursor.execute("""
    UPDATE users
    SET Tên = ?
    WHERE id = ?
""", ("Nguyễn Vân Anh", 1))

# Lưu thay đổi
conn.commit()

print("Tổng số dòng được cập nhật:", conn.total_changes)

# Lấy dữ liệu để kiểm tra
cursor.execute("SELECT * FROM users")
rows = cursor.fetchall()

for row in rows:
    print(row)

# Đóng kết nối
conn.close()
```

```python
import sqlite3

conn = sqlite3.connect("mydatabase.db")
cursor = conn.cursor()

# Tạo bảng nếu chưa tồn tại
cursor.execute("""
CREATE TABLE IF NOT EXISTS users (
    id INTEGER PRIMARY KEY,
    Tên TEXT,
    Tuổi INTEGER
)
""")

# (Tùy chọn) Thêm dữ liệu mẫu nếu bảng đang rỗng
cursor.execute("SELECT COUNT(*) FROM users")
count = cursor.fetchone()[0]

if count == 0:
    cursor.execute("INSERT INTO users (Tên, Tuổi) VALUES (?, ?)", ("Trần A", 20))
    cursor.execute("INSERT INTO users (Tên, Tuổi) VALUES (?, ?)", ("Lê B", 22))
    conn.commit()
    print("Đã thêm dữ liệu mẫu.")

# Cập nhật bản ghi
cursor.execute("""
    UPDATE users
    SET Tên = ?
    WHERE id = ?
""", ("Nguyễn Vân Anh", 1))

conn.commit()

print("Tổng số dòng được cập nhật:", conn.total_changes)

# In tất cả dữ liệu trong bảng
cursor.execute("SELECT * FROM users")
for row in cursor.fetchall():
    print(row)

conn.close()
```
